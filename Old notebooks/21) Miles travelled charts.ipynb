```python
# How many cars on the road?
df['2020 Q3 UK'].sum() / 1_000_000
```

```python
print('total teslas')
print(df[df['Make'] == 'TESLA']['2020 Q3 UK'].sum())

print('teslas sold in 2020')
print(df[df['Make'] == 'TESLA']['2020 Q3 UK'].sum() - df[df['Make'] == 'TESLA']['2019 Q4 UK'].sum())
```

```python
print('total zoes')
print(df[df['Make'] == 'RENAULT'][df[df['Make'] == 'RENAULT']['Model 1'].str.contains('ZOE')]['2020 Q3 UK'].sum())

print('Zoes sold in 2020')
print(df[df['Make'] == 'RENAULT'][df[df['Make'] == 'RENAULT']['Model 1'].str.contains('ZOE')]['2020 Q3 UK'].sum() - df[df['Make'] == 'RENAULT'][df[df['Make'] == 'RENAULT']['Model 1'].str.contains('ZOE')]['2019 Q4 UK'].sum())
```

```python
print('total NISSAN LEAF')
print(df[df['Make'] == 'NISSAN'][df[df['Make'] == 'NISSAN']['Model 1'].str.contains('LEAF')]['2020 Q3 UK'].sum())
print('LEAFs sold in 2020')
print(df[df['Make'] == 'NISSAN'][df[df['Make'] == 'NISSAN']['Model 1'].str.contains('LEAF')]['2020 Q3 UK'].sum() - df[df['Make'] == 'NISSAN'][df[df['Make'] == 'NISSAN']['Model 1'].str.contains('LEAF')]['2019 Q4 UK'].sum())
```
